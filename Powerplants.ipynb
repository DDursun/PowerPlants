```python
df_generation2=generation2.groupby('DATE_TIME').sum().reset_index()
df_generation2['time']=df_generation2['DATE_TIME'].dt.time

fig,ax = plt.subplots(ncols=2,nrows=1,dpi=100,figsize=(20,5))
# daily yield plot
df_generation1.plot(x='DATE_TIME',y='DAILY_YIELD',color='red',ax=ax[0])
# AC & DC power plot
df_generation1.set_index('time').drop('DATE_TIME',1)[['AC_POWER','DC_POWER']].plot(style='o',ax=ax[1])

ax[0].set_title('Daily yield',)
ax[1].set_title('AC & DC power during day hours')
ax[0].set_ylabel('kW',color='red',fontsize=15)
plt.show()
```

```python
df_generation1=generation1.groupby('DATE_TIME').sum().reset_index()
df_generation1['time']=df_generation1['DATE_TIME'].dt.time

fig,ax = plt.subplots(ncols=2,nrows=1,dpi=100,figsize=(20,5))
# daily yield plot
df_generation1.plot(x='DATE_TIME',y='DAILY_YIELD',color='red',ax=ax[0])
# AC & DC power plot
df_generation1.set_index('time').drop('DATE_TIME',1)[['AC_POWER','DC_POWER']].plot(style='o',ax=ax[1])

ax[0].set_title('Daily yield',)
ax[1].set_title('AC & DC power during day hours')
ax[0].set_ylabel('kW',color='red',fontsize=15)
plt.show()
```

```python
daily_generation1=df_generation1.copy()
daily_generation1['date']=daily_generation1['DATE_TIME'].dt.date

daily_generation1=daily_generation1.groupby('date').sum()

fig,ax= plt.subplots(ncols=2,dpi=100,figsize=(20,5))
daily_generation1['DAILY_YIELD'].plot(ax=ax[0],color='navy')
daily_generation1['TOTAL_YIELD'].plot(kind='bar',ax=ax[1],color='navy')
fig.autofmt_xdate(rotation=45)
ax[0].set_title('Daily Yield')
ax[1].set_title('Total Yield')
ax[0].set_ylabel('kW',color='navy',fontsize=17)
plt.show()
```

```python
generation1['day']=generation1['DATE_TIME'].dt.date
generation1=generation1.groupby('day').sum()
generation1['losses']=generation1['AC_POWER']/generation1['DC_POWER']*100

generation1['losses'].plot(style='o--',figsize=(17,5),label='Real Power')

plt.title('% of DC power converted in AC power',size=17)
plt.ylabel('DC power converted (%)',fontsize=14,color='red')
plt.axhline(generation1['losses'].mean(),linestyle='--',color='gray',label='mean')
plt.legend()
plt.show()
```
